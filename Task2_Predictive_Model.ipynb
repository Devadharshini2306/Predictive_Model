#2. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

#3. Upload Dataset

```python
from google.colab import files
uploaded = files.upload()
```

#4. Load Dataset

```python
df = pd.read_csv("train.csv")
df.head()
```

#5. Data Exploration

```python
df.info()
```

```python
df.describe()
```

#6. Handle Data

```python
df = df.select_dtypes(include=['int64', 'float64'])
```

```python
df.head()
```

#7. Handle Missing Values

```python
df = df.fillna(df.mean())
```

#8. Feature Selection

```python
X = df.drop("SalePrice", axis=1)
y = df["SalePrice"]
```

#9. Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

#10. Train Model

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

#11. Prediction

```python
predictions = model.predict(X_test)
```

#12. Model Evaluation

```python
mse = mean_squared_error(y_test, predictions)
print("Mean Squared Error:", mse)
```

#13. Visualization

```python
plt.scatter(y_test, predictions)

plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")

plt.title("Actual vs Predicted Prices")

plt.show()
```

#14. Insights

1. The model predicts house prices using numerical features.
2. Missing values were handled using mean filling.
3. The predicted prices are close to actual values.
4. The model performs reasonably well for basic prediction.
